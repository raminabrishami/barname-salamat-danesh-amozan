```python
import statistics

sen1=0
ghad1=0
vazn1=0
sen2=0
ghad2=0
vazn2=0


class student:
    def __init__(self,tedad,sen,ghad,vazn):
        self.tedad=tedad
        self.sen=sen
        self.ghad=ghad
        self.vazn=vazn
    

class A(student):
    def get_avrage(Self):
        print(Self.sen)
        print(Self.ghad)
        print(Self.vazn)

class B(student):
    def get_avrage(Self):
        print(Self.sen)
        print(Self.ghad)
        print(Self.vazn)


tedad1=int(input('tedad danesh amoz A : '))
z=[float(x) for x in input('sen : ').split()]
sen1=statistics.mean(z)

z1=[float(x) for x in input('ghad :').split()]
ghad1=statistics.mean(z1)

z2=[float(i) for i in input('vazn : ').split()]
vazn1=statistics.mean(z2)




tedad2=int(input('tedad danesh amoz B : ))
f=(float(i) for i in input('sen : ').split())
sen2=statistics.mean(f)

f1=(float(i) for i in input('ghad :').split())
ghad2=statistics.mean(f1)

f2=(float(i) for i in input('vazn : ').split())
vazn2=statistics.mean(f2)


avr=A(tedad1,sen1,ghad1,vazn1)
avr.get_avrage()
avr1=B(tedad2,sen2,ghad2,vazn2)
avr1.get_avrage()

if ghad1>ghad2:
    print('A')
elif ghad2>ghad1:
    print('B')
elif ghad1==ghad2:
    if vazn1>vazn2:
        print('A')
    elif vazn2>vazn1:
        print('B')
    elif vazn1==vazn2:
        print('Same')


    
```
